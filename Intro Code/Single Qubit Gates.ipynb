```python
# s gate: phase gate with phi = pi/2
# also called the square root z gate because ss = z
# note that we can have the hermitian conjugate of the the s gate to get the s-dagger gate

qc = QuantumCircuit(1)
qc.s(0)   
qc.sdg(0)
qc.draw('mpl')
```

```python
# t gate: phase gate with phi = pi/4
# also called the fourth root z gate because tttt = z
# note that we can have the hermitian conjugate of the the t gate to get the t-dagger gate
qc = QuantumCircuit(1)
qc.t(0)   
qc.tdg(0)
qc.draw('mpl')
```

```python
# measuring in a different basis
# computational basis states are eigenstates of the Z gate
# so when we measure in the computational basis we are measuring in the Z basis



# to measure in the X basis we can do this for example
# this is because in the X basis |+> and |-> are the 

def x_measurement(qc, qubit, cbit):
    # Measure 'qubit' in the X-basis  and store the result in 'cbit'
    qc.h(qubit)
    qc.measure(qubit, cbit)
    return qc
```

```python
# trial 1

initial_state = [1/sqrt(2), -1/sqrt(2)]
qc = QuantumCircuit(1,1)
qc.initialize(initial_state, 0)
x_measurement(qc, 0, 0) 
qc.draw()
qobj = assemble(qc) 
counts = sim.run(qobj).result().get_counts()  
plot_histogram(counts)

# makes sense because "|0>" is actually |+> and "|1>" is actually |->
```

```python
# trial 2

initial_state = [1, 0]
qc = QuantumCircuit(1,1)
qc.initialize(initial_state, 0)
x_measurement(qc, 0, 0) 
qc.draw()
qobj = assemble(qc) 
counts = sim.run(qobj).result().get_counts()  
plot_histogram(counts)

# makes sense because "|0>" is actually |+> and "|1>" is actually |->
# so we start with a superposition here and hence we have this
```
